```python
import pandas as pd
import numpy as np
import time
```

```python
#https://newtechaudit.ru/oczenka-kachestva-raboty-sistem-speech-to-text/

def levenshtein_distance(r, h):
    
    rows = len(r)+1
    cols = len(h)+1
    distance = np.zeros((rows,cols),dtype = int)

    for i in range(1, rows):
        for k in range(1,cols):
            distance[i][0] = i
            distance[0][k] = k
   
    for col in range(1, cols):
        for row in range(1, rows):
            if r[row-1] == h[col-1]:
                cost = 0 # если в позиции [i,j] один и тот же символ, то стоимость  = 0
            else:
                cost = 1
            distance[row][col] = min(distance[row-1][col] + 1,      # стоимость deletions
                                 distance[row][col-1] + 1,          # стоимость insertions
                                 distance[row-1][col-1] + cost)     # стоимость substitutions

    return distance[row][col]

def wer(r, h):
    return levenshtein_distance(r.split(), h.split()) / len(r.split()) * 100

def cer(r, h):
    return levenshtein_distance(r, h) / len(r) * 100
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="jonatasgrosman/exp_w2v2t_ru_vp-es_s729")
```

```python

df_dir = 'dataset/Open Russian text/'
df_name = df_dir + 'rez_dataset_open_Russian_text.csv'
df = pd.read_csv(df_name, index_col=0) #list_rez, columns=['text','audio']
df.head()
```

```python
resh = []
count = 0
t = []
for name_par in  range(1,df.shape[0]):
    #print(name_par, 'dataset/dusha/podcast_train/'+path)
    text, path = df.iloc[name_par, :]
    #print(df_dir + path)
    # try:
    if 1:
        t1 = time.time()
        r = pipe(df_dir + path)['text']
        dist_r_text = levenshtein_distance(r, text)
        resh += [[path, r, text, dist_r_text]]
        # print(r, anno) 
        t.append(time.time() - t1)
        
        
    # except:
        count += 1
        if count > 1000:
            break
np.mean(t)     
```
